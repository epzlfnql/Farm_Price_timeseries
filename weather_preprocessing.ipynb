```python
weather_list = glob('/content/drive/MyDrive/농산물예측/aT_data/aT_train_raw/weather*.csv')
weather_list
weather = pd.DataFrame()

for i in weather_list:
  
  tmp_df = pd.read_csv(i)

  # key error나서 잠시 따로 떼어내기
  습도info = tmp_df['습도(%)']
  tmp_df = tmp_df.drop('습도(%)', axis=1)



  
  tmp_df['월&일'] = tmp_df['datadate'].astype('str').apply(lambda x: x[4:]) 
  

  ttmp = tmp_df.groupby('월&일').mean()
  tmp_df = pd.merge(tmp_df, ttmp, how='left', on='월&일')
  
  # 결측치 채우기
  tmp_df['초기온도(℃)_x'] = np.where(pd.notnull(tmp_df['초기온도(℃)_x'])==True, tmp_df['초기온도(℃)_x'], tmp_df['초기온도(℃)_y'])
  tmp_df['최대온도(℃)_x'] = np.where(pd.notnull(tmp_df['최대온도(℃)_x'])==True, tmp_df['최대온도(℃)_x'], tmp_df['최대온도(℃)_y'])
  tmp_df['최저온도(℃)_x'] = np.where(pd.notnull(tmp_df['최저온도(℃)_x'])==True, tmp_df['최저온도(℃)_x'], tmp_df['최저온도(℃)_y'])
  tmp_df['평균온도(℃)_x'] = np.where(pd.notnull(tmp_df['평균온도(℃)_x'])==True, tmp_df['평균온도(℃)_x'], tmp_df['평균온도(℃)_y'])
  tmp_df['강수량(ml)_x'] = np.where(pd.notnull(tmp_df['강수량(ml)_x'])==True, tmp_df['강수량(ml)_x'], tmp_df['강수량(ml)_y'])
  

  tmp_df['습도(%)'] = 습도info
  tmp_df['습도(%)'] = tmp_df['습도(%)'].apply(lambda x: 0 if x==' ' else x)
  # 습도 nan값 0으로 채워넣기
  tmp_df['습도(%)'].fillna(0, inplace=True)
  # 습도 타입 float 형으로
  tmp_df['습도(%)'] = tmp_df['습도(%)'].astype('float')

  
  tmp_df = tmp_df[['datadate_x',	'초기온도(℃)_x',	'최대온도(℃)_x',	'최저온도(℃)_x'	,'평균온도(℃)_x'	,'강수량(ml)_x' ,'습도(%)']]


  # 품목, 주산지 찾는 코드
  found = re.search('weather_(.+?).csv', i).group(1)
  품목, 주산지 = found.split('_')
  tmp_df['품목'] = 품목
  tmp_df['주산지'] = 주산지
  weather = pd.concat([weather, tmp_df], axis=0)


weather
```

```python
# 품목 넣는 코드
final_weather = pd.DataFrame()
for i in range(37): # 품목별
  tmp = pd.DataFrame()
  주산지_arr = weather[weather['품목']==str(i)]['주산지'].unique()
  for j in 주산지_arr: # 품목의 주산지별
    

    품목_df = weather[weather['품목']==str(i)]
    tmp2 = pd.DataFrame()
    
    tmp2['주산지_초기온도'] = 품목_df[품목_df['주산지']==j]['초기온도(℃)_x']
    tmp2['주산지_최대온도'] = 품목_df[품목_df['주산지']==j]['최대온도(℃)_x']
    tmp2['주산지_최저온도'] = 품목_df[품목_df['주산지']==j]['최저온도(℃)_x']
    tmp2['주산지_평균온도'] = 품목_df[품목_df['주산지']==j]['평균온도(℃)_x']
    tmp2['주산지_강수량(ml)'] = 품목_df[품목_df['주산지']==j]['강수량(ml)_x']
    tmp2['습도(%)'] = 품목_df[품목_df['주산지']==j]['습도(%)']
    
    tmp = pd.concat([tmp, tmp2], axis=1)
    
  

   # 주산지가 2개만 들어갈 때
  if len(tmp.columns)<15:
    tmp3 = pd.DataFrame()
    
    tmp3['주산지_초기온도'] = np.nan
    tmp3['주산지_최대온도'] = np.nan
    tmp3['주산지_최저온도'] = np.nan
    tmp3['주산지_평균온도'] = np.nan
    tmp3['주산지_강수량(ml)'] = np.nan
    tmp3['습도(%)'] = np.nan

    tmp = pd.concat([tmp, tmp3], axis=1)
  
  tmp['품목'] = [i]*len(tmp)
  
  final_weather = pd.concat([final_weather, tmp], axis=0)
  

final_weather.columns = ['주산지0_초기온도', '주산지0_최대온도', '주산지0_최저온도', '주산지0_평균온도', '주산지0_강수량(ml)', '주산지0_습도(%)',
       '주산지1_초기온도', '주산지1_최대온도', '주산지1_최저온도', '주산지1_평균온도', '주산지1_강수량(ml)','주산지1_습도(%)',
       '주산지2_초기온도', '주산지2_최대온도', '주산지2_최저온도', '주산지2_평균온도', '주산지2_강수량(ml)', '주산지2_습도(%)','품목']





final_weather['주산지2_초기온도'] = np.where(pd.notnull(final_weather['주산지2_초기온도'])==True, final_weather['주산지2_초기온도'], (final_weather['주산지0_초기온도']+final_weather['주산지1_초기온도'])/2)
final_weather['주산지2_최대온도'] = np.where(pd.notnull(final_weather['주산지2_최대온도'])==True, final_weather['주산지2_최대온도'], (final_weather['주산지0_최대온도']+final_weather['주산지1_최대온도'])/2)
final_weather['주산지2_최저온도'] = np.where(pd.notnull(final_weather['주산지2_최저온도'])==True, final_weather['주산지2_최저온도'], (final_weather['주산지0_최저온도']+final_weather['주산지1_최저온도'])/2)
final_weather['주산지2_평균온도'] = np.where(pd.notnull(final_weather['주산지2_평균온도'])==True, final_weather['주산지2_평균온도'], (final_weather['주산지0_평균온도']+final_weather['주산지1_평균온도'])/2)
final_weather['주산지2_강수량(ml)'] = np.where(pd.notnull(final_weather['주산지2_강수량(ml)'])==True, final_weather['주산지2_강수량(ml)'], (final_weather['주산지0_강수량(ml)']+final_weather['주산지1_강수량(ml)'])/2)
final_weather['주산지2_습도(%)'] = np.where(pd.notnull(final_weather['주산지2_습도(%)'])==True, final_weather['주산지2_습도(%)'], (final_weather['주산지0_습도(%)']+final_weather['주산지1_습도(%)'])/2)

final_weather



```

```python
def my_outlier(pumok, col, vstd = 3.0) :
  #---품목13  주산지1 최저온도
  s = final_weather[(final_weather['품목']==pumok)][col].copy()
  cpcp = pd.DataFrame({"주산지1_최저온도" : s})  

  cpcp['mv5'] = cpcp[col].rolling(window=7).mean() # 20일 이동평균
  cpcp['mv20'] = cpcp[col].rolling(window=28).mean() # 20일 이동평균
  cpcp['st20'] = cpcp[col].rolling(window=28).std() # 20일 이동평균
  cpcp['st5'] = cpcp[col].rolling(window=6).std() # 20일 이동표준편차
  cpcp.fillna(method='bfill', axis=1, inplace=True)
  cpcp['upper'] = cpcp['mv5'] + cpcp['st20']*vstd  # 상단밴드
  cpcp['lower'] = cpcp['mv5'] - cpcp['st20']*vstd # 하단밴드
  cpcp = cpcp[6:] # 20일 이동평균을 구했기 때문에 20번째 행부터 값이 들어가 있음
  
  outlier_idx = cpcp[ ~ ( (cpcp[col] <= cpcp['upper'])   & (cpcp[col] >= cpcp['lower'])) ].index
  cpcp.loc[outlier_idx, col] = cpcp['mv5'] 

  plt.figure(figsize=(9, 5))
  plt.plot(cpcp.index, cpcp[col], label='11')
  plt.plot(cpcp.index, cpcp['mv20'], linestyle='dashed', label='mv20')
  # plt.plot(cpcp.index, cpcp['upper'], linestyle='dashed', label='Upper')
  # plt.plot(cpcp.index, cpcp['lower'], linestyle='dashed', label='lower')
  plt.fill_between(cpcp.index, cpcp['lower'], cpcp['upper'], color='yellow', alpha=0.4)
  plt.legend()
  plt.show()
  plt.close()
```
